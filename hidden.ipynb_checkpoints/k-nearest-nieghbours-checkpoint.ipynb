```python
import joblib
import cv2
import numpy as np
import time
import pyscreenshot as ImageGrab

model = joblib.load("model/digit_recognizer")
knn = cv2.ml.KNearest_create()

train_x_float = np.load("C:/Users/Rumana/Desktop/handwrittenKannada/captured_images/train_x_float.npy")
train_y_encoded = np.load("C:/Users/Rumana/Desktop/handwrittenKannada/captured_images/train_y_encoded.npy")
train_x_float = train_x_float.astype(np.float32)
train_y_encoded = train_y_encoded.astype(np.int32)

knn.train(train_x_float, cv2.ml.ROW_SAMPLE, train_y_encoded)

images_folder = "img/"
accuracy_data = []  # List to store the accuracy results

digit_actual = None  # Variable to store the actual digit drawn by the user

while True:
    img = ImageGrab.grab(bbox=(60, 170, 400, 500))

    img.save(images_folder + "img.png")
    im = cv2.imread(images_folder + "img.png")
    im_gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
    im_gray = cv2.GaussianBlur(im_gray, (15, 15), 0)

    # Threshold the image
    ret, im_th = cv2.threshold(im_gray, 100, 255, cv2.THRESH_BINARY)
    roi = cv2.resize(im_th, (28, 28), interpolation=cv2.INTER_AREA)

    rows, cols = roi.shape

    X = []

    # Add pixel one by one into the data array
    for i in range(rows):
        for j in range(cols):
            k = roi[i, j]
            if k > 100:
                k = 1
            else:
                k = 0
            X.append(k)

    if not any(X):
        digit = "noDigitVisible"
    else:
        predictions = model.predict([X])
        if not predictions[0].isdigit():
            digit = "noDigitVisible"
        else:
            digit = predictions[0]
            
    # Draw prediction and accuracy on the image
    cv2.putText(im, "Prediction: " + str(digit), (20, 20), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 2, cv2.LINE_AA)
    
    cv2.startWindowThread()
    cv2.namedWindow("Result")
    cv2.imshow("Result", im)

    # Wait for user input to set the actual digit
    key = cv2.waitKey(1)
    if key == 27:  # If ESC key is pressed, exit the loop
        break

cv2.destroyAllWindows()
```
