```python
import joblib
import cv2
import numpy as np
import time
import pyscreenshot as ImageGrab

model = joblib.load("model/digit_recognizer")
images_folder = "img/"
accuracy_data = []  # List to store the accuracy results

while True:
    img = ImageGrab.grab(bbox=(60, 170, 400, 500))

    img.save(images_folder + "img.png")
    im = cv2.imread(images_folder + "img.png")
    im_gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
    im_gray = cv2.GaussianBlur(im_gray, (15, 15), 0)

    # Threshold the image
    ret, im_th = cv2.threshold(im_gray, 100, 255, cv2.THRESH_BINARY)
    roi = cv2.resize(im_th, (28, 28), interpolation=cv2.INTER_AREA)

    rows, cols = roi.shape

    X = []

    # Add pixel one by one into the data array
    for i in range(rows):
        for j in range(cols):
            k = roi[i, j]
            if k > 100:
                k = 1
            else:
                k = 0
            X.append(k)

    if not any(X):
        digit = "noDigitVisible"
        accuracy_data.append(0)  # Store incorrect prediction for accuracy calculation
    else:
        predictions = model.predict([X])
        if not predictions[0].isdigit():
            digit = "noDigitVisible"
            accuracy_data.append(0)  # Store incorrect prediction for accuracy calculation
        else:
            digit = predictions[0]
            accuracy_data.append(1)  # Store correct prediction for accuracy calculation

    # Calculate and display accuracy
    total_digits = len(accuracy_data)
    correct_predictions = sum(accuracy_data)
    
    accuracy = correct_predictions / total_digits * 100
    accuracy_text = "Accuracy: {:.2f}%".format(accuracy)

    # Draw prediction and accuracy on the image
    cv2.putText(im, "Prediction: " + str(digit), (20, 20), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 2, cv2.LINE_AA)
    cv2.putText(im, accuracy_text, (20, 50), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 2, cv2.LINE_AA)

    cv2.startWindowThread()
    cv2.namedWindow("Result")
    cv2.imshow("Result", im)

    if cv2.waitKey(1) == 13:  # 27 is the ASCII value of ESC, 13 is the ASCII value of Enter
        break

cv2.destroyAllWindows()
```
